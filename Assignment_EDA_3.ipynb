```python
visa_df=pd.read_csv(r"Visadataset - Visadataset.csv")
d={"Denied":0,"Certified":1}
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df
```

```python
d2={"Bachelor's":0,'Doctorate':1,'High School':2,"Master's":3}
visa_df['education_of_employee']=visa_df['education_of_employee'].map(d2)
visa_df
```

```python
visa_df['full_time_position'].unique()
```

```python
d3={'N':0,'Y':1}
visa_df['full_time_position']=visa_df['full_time_position'].map(d3)
visa_df
```

```python
visa_df['has_job_experience'].unique()
```

```python
visa_df['has_job_experience']=visa_df['has_job_experience'].map(d3)
visa_df
```

```python
visa_df['requires_job_training'].unique()
```

```python
visa_df['requires_job_training']=visa_df['requires_job_training'].map(d3)
visa_df
```

```python
visa_df['region_of_employment'].unique()
```

```python
d4={'Island':0,'Midwest':1,'Northeast':2,'South':3,'West':4}
visa_df['region_of_employment']=visa_df['region_of_employment'].map(d4)
visa_df
```

```python
visa_df['unit_of_wage'].unique()
```

```python
d5={'Hour':0,'Month':1,'Week':2,'Year':3}
visa_df['unit_of_wage']=visa_df['unit_of_wage'].map(d5)
visa_df
```

```python
visa_df=pd.read_csv(r"Visadataset - Visadataset.csv")
cols=visa_df.select_dtypes(include='object').columns
for i in cols:
    lebals=list(visa_df[i].unique())
    values=[i for i in range(len(lebals))]
    d=dict(zip(lebals,values))
    visa_df[i]=visa_df[i].map(d)
visa_df
```

# np.where

```python
visa_df=pd.read_csv(r"Visadataset - Visadataset.csv")
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

```python
con1=visa_df['has_job_experience']=='Y'
visa_df['has_job_experience']=np.where(con1,0,1)
visa_df
```

```python
con1=visa_df['requires_job_training']=='Y'
visa_df['requires_job_training']=np.where(con1,0,1)
visa_df
```
