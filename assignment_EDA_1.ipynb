```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv("Visadataset - Visadataset.csv")
visa_df
```

```python
study=visa_df["education_of_employee"].unique()
study_list=[]
for i in study:
    st=visa_df["education_of_employee"]==i
    val=len(visa_df[st])
    study_list.append(val)
study_list
```

```python
study_df=pd.DataFrame(zip(study,study_list),
                     columns=["Education","Count"])
study_df
```

```python
edf=visa_df['education_of_employee'].value_counts()
```

```python
edf.keys()
```

```python
edf.values
```

### BAR CHART (Education)

**Method-1**

```python
plt.figure(figsize=(6,4))
plt.bar("Education","Count",data=study_df)
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
edf=visa_df['education_of_employee'].value_counts()
keys=edf.keys()
sns.countplot(data=visa_df,x='education_of_employee',order=keys)
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

**Method-2**

```python
edf=visa_df ['education_of_employee'].value_counts()
ax=edf.plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

**Method-3**

```python
plt.figure(figsize=(11,3))
plt.subplot(1,2,1)
edf=visa_df['education_of_employee'].value_counts()
keys=edf.keys()
sns.countplot(data=visa_df,x='education_of_employee',order=keys)
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.subplot(1,2,2)
edf=visa_df ['education_of_employee'].value_counts()
ax=edf.plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Bar chart")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

### PIE CHART

```python
visa_df['education_of_employee'].value_counts(normalize=True)
```

```python
edf=visa_df["education_of_employee"].value_counts()
keys=edf.keys()
values=edf.values
keys,values
```

```python
plt.pie(values,
        explode=[0.1,0.1,0.1,0.1],
        labels=keys,
        autopct="%0.2f%%",
        startangle=180,
        radius=2,
       shadow=1)
plt.show()
```

### Bar Chart (Employ)
