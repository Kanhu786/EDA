```python
import numpy as np
import matplotlib.pyplot as plt
exp_data= np.random.exponential(size=10000)
exp_data
```

### plot the data

```python
plt.hist(exp_data,bins=20,label='Exponential')
plt.show()
```

```python
norm_data= np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Exponential')
plt.show()
```

### Log transfomation

- Log transformation means performing logarithm operations on origianl data

- 
It is an approach to convert into Noraml distributio

- 
Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logarithm')
plt.legend()
plt.show()
```

### Reciprocal Transformations

- 
Reciprocal will fail when values are l

- w
For examples x=0 then 1/0 fail

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label="Reciprocal")
plt.show()
```

### Square root transformation

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=30,label="Square")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=30,label='Logarithm')
plt.legend()
plt.subplot(1,3,3).hist(square_root_data,bins=30,label="Square")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans = power.fit_transform(exp_data1)
plt.hist(data_trans,bins=20,label='Yeo-johnson')
plt.legend()
plt.show()
```

```python
power1 = PowerTransformer(method='box-cox')
exp_data1=exp_data.reshape(-1,1)
data_trans1 = power1.fit_transform(exp_data1)
plt.hist(data_trans1,bins=20,label='Box-Cox')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1).hist(data_trans,bins=20,label='Yeo-johnson')
plt.legend()
plt.subplot(1,2,2).hist(data_trans1,bins=20,label='Box-Cox')
plt.legend()
plt.show()
```
